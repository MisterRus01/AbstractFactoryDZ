```python
for column in dataset.select_dtypes(include=[np.number]).columns:
    sns.histplot(dataset[column])
    plt.title(f'Distribucija {column}')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
binary_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for column in binary_columns:
    dataset[column] = le.fit_transform(dataset[column])


from sklearn.preprocessing import OneHotEncoder
categorical_columns = ['furnishingstatus']
dataset = pd.get_dummies(dataset, columns=categorical_columns)
dataset.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
for column in dataset.select_dtypes(include=[np.number]).columns:
    vif = variance_inflation_factor(dataset.select_dtypes(include=[np.number]).values, dataset.columns.get_loc(column))
    print(f'VIF za {column}: {vif}')

```

```python
for column in dataset.columns:
    if dataset[column].dtype in ['int64', 'float64']:  # Provjera je li stupac numerički
        mean = dataset[column].mean()  # Srednja vrijednost
        std = dataset[column].std()   # Standardna devijacija
        dataset[column] = (dataset[column] - mean) / std
dataset.head()
```

```python
X = dataset[['area', 'price']]
y = dataset['bedrooms']

from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X, y)
import statsmodels.api as sm

X = sm.add_constant(X)  # Dodavanje intercepta
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
residuals = model.resid
sns.histplot(residuals, kde=True)
plt.title('Distribucija reziduala')
plt.show()
```

```python
predictions = model.predict(X)
plt.scatter(y, predictions)
plt.xlabel('Stvarne vrijednosti')
plt.ylabel('Predviđene vrijednosti')
plt.title('Stvarne vs. Predviđene vrijednosti')
plt.show()
plt.scatter(predictions, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predviđene vrijednosti')
plt.ylabel('Rezidualne vrijednosti')
plt.title('Rezidualni graf')
plt.show()
```
