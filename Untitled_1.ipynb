```python
import seaborn as sns
import pandas as pd
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
data = pd.read_csv('processed_data.csv') #ucitavanje podataka
data = data.dropna() #odbacivanje nedostajućih vrijednosti
clearedData = data.drop_duplicates() # uklanjanje dupliciranih vrijednosti
print(clearedData.head()) #pregled prvih 5 redaka
```

```python
print(clearedData.info()) #prikaz tipova varijabli
```

```python
print(clearedData.describe()) #opisna statistika za numeričke vrijednosti
#for column in data:
 #   print(data[column].unique())
```

```python
print(clearedData.count())
print(clearedData.shape) # broj redaka | broj stupaca
```

```python
# 6. Grafička analiza - histogrami
plt.figure(figsize=(8, 4))
plt.hist(data['currency'], bins=20, edgecolor='black')
plt.title(f"Histogram za ponavljanje valute")
plt.xlabel('Currency')
plt.ylabel('Frekvencija')
plt.show()
#Najčešća valuta je USD, no nije velika razlika između tri moguće valute

plt.figure(figsize=(8, 4))
plt.hist(data['price_USD'], bins=20, edgecolor='black')
plt.title(f"Histogram za cijene")
plt.xlabel('Cijena USD')
plt.ylabel('Frekvencija')
plt.show()
#Naviše ponavljana cijena je između 200 i 500, kako se cijena povečava, manje je mobitela

plt.figure(figsize=(8, 4))
plt.hist(data['BATTERY'], bins=20, edgecolor='black')
plt.title(f"Histogram za ponavljanje kapaciteta baterije")
plt.xlabel('Battery[MAh]')
plt.ylabel('Frekvencija')
plt.show()
#Većina mobitela sadrži između 4000 - 6000MAh kapaciteta baterije
```

```python
sns.boxplot(x=clearedData['BATTERY'])
plt.title('Boxplot - BATTERY')
plt.show()
#prikaz kapaciteta baterije boxplotom

sns.boxplot(x=clearedData['Weight'])
plt.title('Boxplot - Weight')
plt.show()
#prikaz težine telefona boxplotom
```

```python
skewness = skew(clearedData['BATTERY'])
kurt = kurtosis(clearedData['BATTERY'])
print('Skewness:', skewness)
print('Kurtosis:', kurt)
print("")
skewness = skew(clearedData['Year'])
kurt = kurtosis(clearedData['Year'])
print('Skewness:', skewness)
print('Kurtosis:', kurt)
```

```python
sns.countplot(x='phone_brand', data=data)
plt.title('Distribucija - Kategorijska Kolona')
plt.show()
#Histogram distribucije phone brand
```

```python
plt.scatter(clearedData["Weight"], clearedData["BATTERY"])
#povečanjem težine se uglavnom povećava i kapacitet baterije
```

```python
#isto povećanjem težine se uglavnom povečava i veličina zaslona
plt.scatter(clearedData["Display_Size"], clearedData["Weight"])
```

```python
sns.countplot(clearedData["store"])
#distribucija za trgovine, očekivano se najviše prodaje na amazon stranicama, dok na ostalima po dosta manje
```

```python
#heatmap za sve numeričke varijable
corr_matrix = clearedData.select_dtypes(include=['float64', 'int64']).drop(columns=['year'], errors='ignore').corr()
print("\nKorelacijski matriks:")
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Korelacijski matriks - Phone Dataset')
plt.show()
```

```python
print("\nDeskriptivna Statistika po brendu mobitela:")
print(clearedData.groupby('phone_brand').describe())
#-mean prosjek, std- standardna devijacija, minimunm, prvi kvartal, drugi kvartal
```
