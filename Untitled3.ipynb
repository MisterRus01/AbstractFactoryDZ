c) U oba slucaja, kada smo koristili razinu značajnosti 5% i 1% smo dobili da su vrijendosti veće od P vrijednosti korištenog testa. Time zaključujemo da možemo odbaciti nultu hipotezu i imamo dovoljno dokaza da je srednja vrijednost težine papira razlicita u oba slucaja.

```python
from math import sqrt
n1 = np.size(papir1)
n2 = np.size(papir2)

x1 = np.mean(papir1)
x2 = np.mean(papir2)

s1 = np.std(papir1)
s2 = np.std(papir2)

sp = sqrt(((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 ** 2) / (n1 + n2 - 2))

t= n1+n2-2
print(t)
t=2.048
up = (x1-x2) + t * sp * sqrt(1/n1 + 1/n2)
down = (x1-x2) - t * sp * sqrt(1/n1 + 1/n2)
print("[" + str(down) + ", " + str(up) + "]")

t = 2.763
up = (x1-x2) + t * sp * sqrt(1/n1 + 1/n2)
down = (x1-x2) - t * sp * sqrt(1/n1 + 1/n2)
print("[" + str(down) + ", " + str(up) + "]")
```
